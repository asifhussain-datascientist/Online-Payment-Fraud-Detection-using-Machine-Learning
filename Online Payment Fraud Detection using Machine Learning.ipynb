**Data Preprocessing**

This step includes the following:
* Encoding of Typecolumn
* Dropping irrelevant columns like nameOrig, nameDest
* Data Splitting

```python
type_new = pd.get_dummies(data['type'], drop_first=True)
data_new = pd.concat([data, type_new], axis=1)
data_new.head()
```

Once we done with the encoding, now we can drop the irrelevant columns.
For that, follow the code given below.

```python
X = data_new.drop(['isFlaggedFraud', 'isFraud','type', 'nameOrig', 'nameDest'], axis=1)
y = data_new['isFlaggedFraud']
```

**let's check the shape of extracted data.**

```python
X.shape, y.shape
```

**now let's split the data into two parts : Training and Testing**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

**Model Training**
As the prediction is a classification problem so the models we will be using are :

* LogisticRegression: It predicts that the probability of a given data belongs to the particular category or not.
* XGBClassifier: It refers to Gradient Boosted deicion trees. In this algorithm, desicion trees are created in sequential from the weights are assigned to all the independent variables which are then fed into the decision tree which predicts results.
* SVC: SVC is used to find a hyperplae in an N-dimentional space that distinctly classifies the data points. Then it gives the output according the most nearby element.
* RandomForestClassifier: Random forest classifier creates a set of decision tres from a randomly selected subset of the training set. Then, it collects the votes from different decision trees to decide the final prediction.

**Let's import the modules of the relevant models.**

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score as ras
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

**Let's train the model**

```python
models = [LogisticRegression(), XGBClassifier(),
          RandomForestClassifier(n_estimators=7,
                                 criterion='entropy', 
                                 random_state=7)]

for i in range(len(models)):
    models[i].fit(X_train, y_train)
    print(f'{models[i]} : ')
    
    train_preds = models[i].predict_proba(X_train)[:, 1]
    print('Training Accuracy : ', ras(y_train, train_preds))
    
    y_preds = models[i].predict_proba(X_test)[:, 1]
    print('Validation Accuracy : ', ras(y_test, y_preds))
    print()
```

**Model Evaluation**

The best performed model is XGBClassifier. Let's plot the Confusion Matrix for the same

```python
from sklearn.metrics import ConfusionMatrixDisplay

cm = ConfusionMatrixDisplay.from_estimator(models[1], X_test, y_test) 

cm.plot(cmap='Blues') 

plt.show()
```
